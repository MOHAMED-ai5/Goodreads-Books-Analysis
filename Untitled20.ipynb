```python
!pip install pandas matplotlib seaborn -q


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Goodreads_books_with_genres.csv")
print("عدد الصفوف والأعمدة:", df.shape)
df.head()



df.drop_duplicates(inplace=True)


df.dropna(subset=['average_rating', 'num_pages', 'ratings_count'], inplace=True)


print("\nإحصائيات التقييم:")
print(df['average_rating'].describe())

top_authors = df['Author'].value_counts().head(10)


top_books = df.sort_values(by='ratings_count', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.histplot(df['average_rating'], bins=30, kde=True, color='purple')
plt.title('Distribution of ratings', fontsize=16)
plt.xlabel('evaluation')
plt.ylabel('number of books')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x=top_authors.values, y=top_authors.index, palette="viridis")
plt.title('Most Prolific Authors', fontsize=16)
plt.xlabel('number of books')
plt.ylabel('author')
plt.show()
print("                     ")
print("                     ")
plt.figure(figsize=(10,6))
sns.scatterplot(x='num_pages', y='average_rating', data=df, alpha=0.5, color='orange')
plt.title('The relationship between the number of pages and the average rating', fontsize=16)
plt.xlabel('number of pages')
plt.ylabel('Average rating')
plt.show()


print("\n📊 مشروع تحليل بيانات كتب Goodreads")
print(f"✅ عدد الكتب: {len(df)}")
print(f"✅ أعلى كتاب تقييماً: {df.loc[df['average_rating'].idxmax(), 'Title']}")
print(f"✅ المؤلف الأكثر إنتاجاً: {top_authors.index[0]}")

```
